### Track Duration

```python

# Define the function to convert duration
def convert_duration(milliseconds):
    seconds = milliseconds // 1000
    minutes = seconds // 60
    seconds = seconds % 60
    return f"{minutes:02}:{seconds:02}"

# Calculate average duration per artist
average_duration_per_artist = df.groupby('Artist')['Duration'].mean().reset_index()

# Convert the average duration from milliseconds to mm:ss format
average_duration_per_artist['Average Duration (mm:ss)'] = average_duration_per_artist['Duration'].apply(lambda x: convert_duration(int(round(x))))

#Original Data:
# A lambda function is an anonymous, inline function defined with the lambda keyword.
# df['Duration'] might be [123456, 654321].
# Apply lambda Function:
#First Element:
#x = 123456
#round(123456) → 123456 (no change) round(x) rounds the value of x to the nearest integer
#int(123456) → 123456
#convert_duration(123456) → "02:03" (assuming 123456 ms converts to 2 minutes and 3 seconds)

# Select only the relevant columns
average_duration_per_artist = average_duration_per_artist[['Artist', 'Average Duration (mm:ss)']]

print(average_duration_per_artist)
```

# Analyze trends and correlations in df

```python
# calculate the correlation coefficient between song duration and popularity
correlation = df[['Duration', 'Popularity']].corr().iloc[0, 1] #.iloc[] is used for positional indexing. It allows you to select specific values from the DataFrame based on their row and column positions.
print(f'Correlation coefficient between song duration and popularity: {correlation:.2f}')  #:.2f formats the correlation value to two decimal places.
```

The correlation coefficient of -0.13 between song duration and popularity indicates a weak negative relationship. This means that as the duration of the song increases, its popularity tends to slightly decrease, but the relationship is not very strong.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
file_path = 'WorldHits.csv'
df = pd.read_csv(file_path)
df
```

# Histogram 

```python
# List of columns to plot histograms for
columns_to_plot = ['Danceability', 'Energy', 'Loudness', 'Tempo', 'Popularity']

# Create histograms with KDE for each column
for column in columns_to_plot:
    plt.figure(figsize=(10, 6))
    sns.histplot(df[column], bins=30, kde=True)
    plt.title(f'Histogram and KDE of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()



#Most tracks have a danceability score between 0.4 and 0.7, peaking around 0.6.
#Very few tracks have low danceability (<0.2) or high danceability (>0.8).
```

In the context of histograms, KDE is used to overlay a smooth curve on top of the histogram, which can help in understanding the distribution of the data more clearly.

```python
# Count the number of occurrences for each year and get the top 10
year_counts = df['Year'].value_counts().head(10)

# Create a bar plot
year_counts.plot(kind='bar', title='Number of Music Tracks by Year', color='skyblue')
plt.xlabel('Year')
plt.ylabel('Number of Tracks')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
# Create a pivot table with years as rows and artists as columns
pivot_table = pd.crosstab(df['Year'], df['Artist'])

# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, cmap='YlGnBu', annot=True)
plt.title('Heatmap of Number of Music Tracks by Year and Artist')
plt.xlabel('Artist')
plt.ylabel('Year')
plt.show()

# the number of music tracks by different artists over time (years).
```

# Compute the average Danceability for each year

```python
# Compute the average Danceability for each year
average_danceability = df.groupby('Year')['Danceability'].mean().reset_index()

# Create a line plot
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Danceability', data=average_danceability, marker='o')
plt.title('Average Danceability Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Danceability')

plt.show()
```

```python
# Create a scatter plot
plt.figure(figsize=(10, 6))  # Set figure size
sns.scatterplot(x='Danceability', y='Energy', data=df, color='blue', marker='o', s=100, alpha=0.7)
plt.title('Scatter Plot of Danceability vs. Energy')
plt.xlabel('Danceability')
plt.ylabel('Energy')
plt.grid(True)
plt.show()
```

```python


# Select the subset of columns for the pairplot
subset_columns = ['Danceability', 'Energy', 'Tempo', 'Loudness', 'Popularity']
subset_df = df[subset_columns]

# Create a pairplot for the selected columns
sns.pairplot(subset_df)
plt.suptitle('Pairplot of Danceability, Energy, Tempo, Loudness, and Popularity', y=1.02)
plt.show()
```


Observations:
Loudness and Energy appear to be somewhat positively correlated, which aligns with the idea that louder tracks tend to feel more energetic.
Popularity does not show a clear correlation with any of the other features, suggesting that factors beyond Danceability, Energy, Tempo, and Loudness likely contribute to a track's popularity.
Danceability and Energy have a moderate spread, meaning different levels of these attributes are represented in the dataset, but they don't exhibit clear relationships with each other or with other features.
